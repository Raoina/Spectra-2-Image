```python
import pandas as pd

file_path = '/content/drive/MyDrive/DL_FreshCS_train_Multiple_Refs.csv'

try:
  df = pd.read_csv(file_path)
  display(df.head())
except FileNotFoundError:
  print(f"Error: The file '{file_path}' was not found. Please check the path.")
except Exception as e:
  print(f"An error occurred: {e}")
```

```python
# Replace ['column_name_1', 'column_name_2'] with the actual names of the columns you want to drop
columns_to_drop = ['Scanner ID']

# Drop the columns
df.drop(columns=columns_to_drop, axis=1, inplace=True)

# Display the first few rows of the DataFrame after dropping columns
display(df.head())
```

```python
df.shape
```

```python
# Group by 'Scanner ID' and calculate the mean of other numerical columns
averaged_df = df.groupby('Sample ID').mean()

# Reset the index to turn 'Scanner ID' back into a column
averaged_df.reset_index(inplace=True)

print("DataFrame with averaged values grouped by Scanner ID:")
display(averaged_df.head())
```

```python
averaged_df.shape
```

```python
# Replace with the actual names of the columns for the first DataFrame
target = ['Sample ID', 'Moi']

# Replace with the actual names of the columns for the second DataFrame
# You can include the remaining columns or a specific subset
# Include 'Scanner ID' in df2 as well
spectra = [col for col in averaged_df.columns if col not in ['Moi']]


df1 = averaged_df[target].copy()
df2 = averaged_df[spectra].copy()

print("First DataFrame (df1) head:")
display(df1.head())

print("\nSecond DataFrame (df2) head:")
display(df2.head())
```
